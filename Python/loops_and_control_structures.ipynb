```python
# 1.1
marks = int(input("enter marks: "))
if 0 < marks <= 100:
  if 90 <= marks <=100:
    print("you got A grade")
  elif 80 <= marks <=89:
    print("you got B grade")
  elif 70 <= marks <=79:
    print("you got C grade")
  elif 60 <= marks <= 69:
    print("you got D grade")
  else :
    print("you got F grade")
else :
  print("Enter valid marks")
```

```python
#1.2
year = int(input("Enter a year: "))
if year % 4==0 and year % 100 == 0 and year % 400 ==0:
  print(f"{year} is a leap year")
else:
  print(f"{year} year is not a leap year")
```

```python
#1.3
a = int(input("Enter a side of a triangle: "))
b = int(input("Enter a side of a triangle: "))
c = int(input("Enter a side of a triangle: "))
if a+b > c or b+c > a or c+a > b:
  if a==b==c:
    print("It's an Equilateral triangle")
  elif a==b or b==c or c==a :
    print("It's an Isolated triangle")
  else:
    print("It's Scalene triangle")
else:
  print("It's not a valid triangle ")
```

```python
#1.4
weight = float(input("Enter your weight in kgs :"))
height = float(input("Enter your height in m**2 :"))
BMI = weight // height * 2
if BMI <= 18.5:
  print("Your under weight, you need more protein")
elif 18.5 < BMI < 24.9:
  print("Your normal in weight")
elif 25 < BMI < 29.9:
  print("Your overweight do some workouts")
else:
  print("Your obese")
```

Section 3: While Loops

```python
#3.1
num = int(input("enter a number : "))
rev = 0
while num !=0:
  digit = num % 10
  rev = rev*10 + digit
  num = num // 10
print(rev)
```

```python
#3.2
num = int(input("enter a number : "))
sum_digi = 0
while num>0:
  sum_digi = sum_digi + num%10
  num = num // 10
print(sum_digi)
```

```python
#3.4
numm =int(input("enter a number :"))
num = numm
digi = 0
while num!=0:
  rev = num%10
  digi = digi*10 +rev
  num=num//10
if numm == digi :
  print("its a palindrome")
else:
  print("its not")
```

Section 5: Break and Continue Statements

```python
#5.1
for i in range(1,101):
  if i % 7==0 and i%13 == 0:
    print(i)
    break
```

```python
#5.2
for i in range(1,51):
  if i%3==0:
    continue
  else:
    print(i,end=" ")
```

```python
#5.3
total =0
while True:
  n = int(input("enter a number: "))
  if n<0:
    break
  elif n==0:
    continue
  else:
    total+=n
    print(total)
```
