```python
#  Remove all vowels from 'Computer Science'
s = 'Computer Science'
result = ""
for i in s:
  if i.lower() not in "aeiou":
    result = result + i
print(result)
```

```python
# mississippi
n = 'mississippi'
r ={}
for i in n:
  if i in r:
    r[i] += 1
else:
  r[i] = 1
max_r = max(r, key=r.get)
print(max_r)
```

```python
# Check if two strings are anagrams: 'listen' and 'silent'
t = 'listen'
s = 'silent'
if sorted(t) == sorted(s):
  print(True)
else:
  print(False)
```
