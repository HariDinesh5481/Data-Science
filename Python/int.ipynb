```python
# Check if 97 is a prime number
num = 97
if num <= 1:
    print("It is a not a prime number")
else :
    is_prime = True
    for i in range(2, int(num**0.5)+1):
        if num % i ==0:
            is_prime = False
            break
if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# Find the factorial of 8 FOR LOOP
num = 8
factorial = 1
for i in range(1,num+1):
    factorial = factorial * i
print(f"The factorial of {num} is {factorial}")
```

```python
# Find the factorial of 8 WHILE LOOP
num = 8 
fact = 1
i=1
while i <= num:
    fact = fact * i
    i = i +1 
print(f"The factorial of {num} is {fact}")
```

```python
#Calculate the average of numbers: 15, 23, 31, 42, 56
num = (15, 23, 31, 42, 56)
sum = 0
for i in num:
  sum = sum + i
  avg = sum/len(num)
print(avg)
```

```python
# Find the greatest common divisor (GCD) of 48 and 36
a = 60  # first number
b = 48  # second number

# loop until the remainder is 0
while b != 0:
    a, b = b, a % b

print(a)
```
