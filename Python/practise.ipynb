```python
def chec(*numbers):
    max_num =numbers[0]
    min_num =numbers[0]
    for i in numbers:
        if i > max_num:
            max_num=i
        elif i < min_num:
            min_num=i
    return(max_num,min_num)
```

```python
chec(1,2,3,4,5,4)
```
