```python
price = float(input("Enter purchase amount: "))

if price > 100:
    price = price * 0.9      # 10% discount
elif price > 50:
    price = price * 0.95     # 5% discount

print("Final price:", price)
```

```python
#6
year = int(input("Enter a year: "))

if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```python
weight = float(input("Enter package weight (kg): "))
destination = input("Destination (domestic/international): ").lower()
shipping_type = input("Shipping type (standard/express): ").lower()

if destination == "domestic":
    rate = 5
elif destination == "international":
    rate = 15
else:
    print("Invalid destination")
    exit()

if shipping_type == "standard":
    multiplier = 1.0
elif shipping_type == "express":
    multiplier = 1.5
else:
    print("Invalid shipping type")
    exit()

cost = weight * rate * multiplier
print(f"Total Shipping Cost: ${cost:}")
```
