```python
#Square a number using lambda.
v = lambda x:x**2
print(v(2))
```

```python
# Add two numbers using lambda.
add_num = lambda a,b:a+b
print(add_num(2,3))
```

```python
# Find cube of numbers using map + lambda:
c = [1,2,3,4]
cube = list(map(lambda x:x**3,c))
print(cube)

```

```python
#Filter even numbers using lambda:

f = [1,2,3,4,5,6]
filter_even = list(filter(lambda x: x % 2==0,f))
```

```python
filter_even
```

```python
# Convert list to uppercase using lambda:
v = ['hi','hello']
uppper = list(map(lambda x : x.upper(),v))
```

```python
uppper
```

```python
# Find maximum using reduce:
from functools import reduce
r = [10,25,3,90,50]
max_r = reduce(lambda a,b : a if a > b else b ,r)

```

```python
max_r
```

```python
# Sort list of tuples by second element:

d = [(1,3),(2,1),(4,2)]
sort_d = sorted(d,key = lambda x : x[1])
sort_d
```

```python
# Find product of all numbers using reduce.
from functools import reduce
g = (1,2,3,4,5)
product_g = reduce(lambda a,b : a*b,g)
product_g
```

```python
from functools import reduce
h = {'h':1,'J':2,'r':20}
product_h = reduce(lambda a,b :a*b,h.values())
product_h
```

```python
# Filter numbers greater than 10:

b = [5,12,7,18,3]
num_greater = list(filter(lambda x : x > 10 ,b))
```

```python
num_greater
```

```python
# Double only even numbers using lambda + map.
b = [5,12,7,18,3]
double = list(map(lambda x : x*2 if x%2==0 else x,b))
```

```python
double
```

```python
#Find longest word using reduce:
v = ['apple','banana','kiwi']
sorted_v = sorted(v,key=lambda a:len(a))
```

```python
from functools import reduce
v = ['apple','banana','kiwi']
reduced = reduce(lambda x,y : x if len(x) >= len(y) else y , v)
```

```python
reduced
```
