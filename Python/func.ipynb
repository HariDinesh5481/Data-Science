```python
#Write a function that takes a number as input and returns whether it is even or odd.
def check(n):
    if n %2==0:
        return('even')
    else:
        return('odd')
```

```python
check(3)
```

```python
#Write a function that takes a number and returns its factorial.
def fact(n):
    factorial = 1
    for i in range(1,n+1):
        factorial=factorial*i
    return(factorial)
```

```python
fact(3)
```

```python
# Write a function that takes a string and returns the reversed string.
def rev(name):
    return(name[::-1])
```

```python
# Write a function that checks whether a number is prime.
def is_prime(n):
    if n <= 1:
        return False 
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return (True)
```

```python
# Write a function that takes a list of numbers and returns the largest number.
def take(num):
    max_num = num[0]
    for i in num:
        if i > max_num:
            max_num = i
    return(max_num)
```

```python
print(take([1,2,3,4]))
```

```python
# Write a function that takes a list and returns the sum of all elements without using the built-in `sum()` function.
def takes(n):
    sum = 0
    for i in n:
        sum+=i
    return sum


```

```python
print(takes([1,2,3,4]))
```
